```python
import numpy as np

# Matriz de covarianza y matriz de correlación empírica
matriz_cov = np.array([[140.017, 107.881, 139.068, 109.095],
                       [0, 106.038, 110.0439, 82.627],
                       [0, 0, 168.752, 125.136],
                       [0, 0, 0, 108.960]])

matriz_corr = np.array([[.969, -.102, .194, .116],
                        [.907, -.392, -.106, -.111],
                        [.971, .156, -.157, .092],
                        [.943, .252, .092, -.196]])

# (1) Varianzas empíricas de las componentes principales
varianzas_componentes_principales = np.diag(matriz_cov)

# (2) Proporciones de varianza explicadas por cada componente principal
autovalores, _ = np.linalg.eig(matriz_cov )
prop_varianza_explicada_cov = autovalores / np.sum(autovalores)

autovalores_corr, _ = np.linalg.eig(matriz_corr)
prop_varianza_explicada_corr = autovalores_corr / np.sum(autovalores_corr)

# Imprimir resultados
print("(1) Varianzas empíricas de las componentes principales (utilizando la matriz de covarianza):")
print(varianzas_componentes_principales)

print("\n(2) Proporciones de varianza explicadas por cada componente principal (utilizando la matriz de covarianza):")
print(prop_varianza_explicada_cov)

print("\n(3) Proporciones de varianza explicadas por cada componente principal (utilizando la matriz de correlación):")
print(prop_varianza_explicada_corr)
```
